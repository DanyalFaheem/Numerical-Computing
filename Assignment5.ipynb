```python
# Question 1

import sympy
import numpy as np
from copy import deepcopy
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')
X1 = 1/3 - 1/3*(-y + z)
Y1 = -1/6*(3*x + 2*z)
Z1 = 4/3 - 1/7*(3*x + 3*y)

values = {'X': 0, 'Y': 0, 'Z': 0}
previousvalues = {'X': 0, 'Y': 0, 'Z': 0}
currentvalue = []
while True:
    previousvalues['X'] = X1.subs([(y, values['Y']), (z, values['Z'])])
    currentvalue.clear()
    previousvalues['Y'] = Y1.subs([(z, values['Z']), (x, values['X'])])
    currentvalue.clear()
    previousvalues['Z'] = Z1.subs([(y, values['Y']), (x, values['X'])])
    currentvalue.clear()
    newvalues = {key: previousvalues[key] -
                 values.get(key, 0) for key in values}
    max = list(newvalues.values())
    print(max, np.max(max))
    values = deepcopy(previousvalues)
    if np.abs(np.max(np.abs(max))) <= 0.001:
        break
    print(previousvalues)

print("Jacoi Final Solution is: ", values)


values = {'X': 0, 'Y': 0, 'Z': 0}
previousvalues = {'X': 0, 'Y': 0, 'Z': 0}
currentvalue = []
while True:
    previousvalues['X'] = X1.subs(
        [(y, previousvalues['Y']), (z, previousvalues['Z'])])
    currentvalue.clear()
    previousvalues['Y'] = Y1.subs(
        [(z, previousvalues['Z']), (x, previousvalues['X'])])
    currentvalue.clear()
    previousvalues['Z'] = Z1.subs(
        [(y, previousvalues['Y']), (x, previousvalues['X'])])
    currentvalue.clear()
    newvalues = {key: previousvalues[key] -
                 values.get(key, 0) for key in values}
    max = list(newvalues.values())
    print(max, np.max(np.abs(max)))
    values = deepcopy(previousvalues)
    if np.abs(np.max(np.abs(max))) <= 0.001:
        break
    print(previousvalues)

print("Gauss Seidel Final Solution is: ", values)

W = 1.2
X1 = (1-1.2)*x + 1.2*(1/3 - 1/3*(-y + z))
Y1 = (1-1.2)*y + 1.2*(0 - 1/6*(3*x + 2*z))
Z1 = (1-1.2)*z + 1.2*(4/3 - 1/7*(3*x + 3*y))

values = {'X': 0, 'Y': 0, 'Z': 0}
previousvalues = {'X': 0, 'Y': 0, 'Z': 0}
currentvalue = []
while True:
    previousvalues['X'] = X1.subs(
        [(x, previousvalues['X']), (y, previousvalues['Y']), (z, previousvalues['Z'])])
    currentvalue.clear()
    previousvalues['Y'] = Y1.subs(
        [(z, previousvalues['Z']), (y, previousvalues['Y']), (x, previousvalues['X'])])
    currentvalue.clear()
    previousvalues['Z'] = Z1.subs(
        [(y, previousvalues['Y']), (z, previousvalues['Z']), (x, previousvalues['X'])])
    currentvalue.clear()
    newvalues = {key: previousvalues[key] -
                 values.get(key, 0) for key in values}
    max = list(newvalues.values())
    print(max, np.max(np.abs(max)))
    values = deepcopy(previousvalues)
    if np.abs(np.max(np.abs(max))) <= 0.001:
        break
    print(previousvalues)

print("SOR Final Solution is: ", values)
```

```python
# Question 2
import sympy
import numpy as np
from copy import deepcopy
x = sympy.symbols('x')
y = sympy.symbols('y')
X1 = 59.2/0.03 - 1/0.03*(58.9*y)
Y1 = 47/6.1 - 1/6.10*(5.31*x)

values = {'X': 0, 'Y': 0}
previousvalues = {'X': 0, 'Y': 0}
currentvalue = []
n = 0
while n < 15:
    previousvalues['X'] = X1.subs(y, previousvalues['Y'])
    currentvalue.clear()
    previousvalues['Y'] = Y1.subs(x, previousvalues['X'])
    currentvalue.clear()

    newvalues = {key: previousvalues[key] -
                 values.get(key, 0) for key in values}
    max = list(newvalues.values())
    print(max, np.max(max))
    values = deepcopy(previousvalues)
    if np.max(np.abs(max)) <= 0.1:
        break
    print(previousvalues)
    n += 1
print("Final Solution is: ", values)
```
